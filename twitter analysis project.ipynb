# Twitter data




## Authorizing  to access Twitter account data

```python
import twitter
import json
import os
import pickle
```

```python
if not os.path.exists('secret_twitter_credentials1.pkl'):
    Twitter={}
    Twitter['Consumer Key'] = ''
    Twitter['Consumer Secret'] = ''
    Twitter['Access Token'] = ''
    Twitter['Access Token Secret'] = ''
    with open('secret_twitter_credentials.pkl','wb') as f:
        pickle.dump(Twitter, f)
else:
    Twitter=pickle.load(open('secret_twitter_credentials1.pkl','rb'))
```

```python

auth = twitter.oauth.OAuth(Twitter['Access Token'],
                           Twitter['Access Token Secret'],
                           Twitter['Consumer Key'],
                           Twitter['Consumer Secret'])

twitter_api = twitter.Twitter(auth=auth)

print(twitter_api)
```

## Finding the viral trends in twitter 

```python

india_id = 23424848
hyd_id = 2295414
```

```python
india_trends = twitter_api.trends.place(_id = india_id)
india_trends[:1] # confirming that we find or not
```

```python
hyd_trends = twitter_api.trends.place(_id = hyd_id)
#hyd_trends[:1] # confirming that we find or not 
```

## Displaying trends 

```python
trends=hyd_trends
print(type(trends))
print(list(trends[0].keys()))
print(trends[0]['trends'])
```

```python
# displaying above data with more beautiful way by json

print((json.dumps(hyd_trends[:1], indent=2)))
```

## Displaying only trends without meta data

```python
trends_dict = {}
trends_dict['india'] = set([trend['name'] 
                        for trend in india_trends[0]['trends']])

trends_dict['hyd'] = set([trend['name'] 
                     for trend in hyd_trends[0]['trends']]) 
```

```python
for i in ['india','hyd']:
    print(('**'*15,i))
    print((','.join(trends_dict[i])))
```

##  Comparing how much hyderabad trends are related toIndian total trends

```python
print( '---'*10,'intersection of india and hyd')
trends_intersection = trends_dict['india'].intersection(trends_dict['hyd'])

print(trends_intersection)
```

## Finding  tweets by hashtags

```python
q = '#Prabhas' 

number = 100

search_results = twitter_api.search.tweets(q=q, count=number)

statuses = search_results['statuses']

#print(search_results)
```

```python
type(statuses)

#print((json.dumps(statuses, indent=1)))
```

```python
len(statuses)
```

## Twitter gives Duplicates so removing them 

```python
all_text = []
filtered_statuses = []
for s in statuses:
    if not s["text"] in all_text:
        filtered_statuses.append(s)
        all_text.append(s["text"])
statuses = filtered_statuses     
```

```python
# checking the tweets count after removing duplicates
len(statuses)
```
